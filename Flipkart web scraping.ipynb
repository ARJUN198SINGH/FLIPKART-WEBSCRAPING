```python
import bs4
import requests
```

```python
url='https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
data=requests.get(url)
soup=bs4.BeautifulSoup(data.text,'html.parser')
print(soup.prettify)
```

```python
mobile_name=soup.find_all('div',{'class':'_4rR01T'})
mobile_names=[]
for i in mobile_name:
    mobile_names.append(i.text)
```

```python
mobile_rate=soup.find_all('div',{'class':'_3LWZlK'})
mobile_rating=[]
for i in range(len(mobile_names)):
    mobile_rating.append(mobile_rate[i].text)
```

```python
mobile_off=soup.find_all('div',{'class':'_3Ay6Sb'})
mobile_discount=[]
for i in range(len(mobile_names)):
    mobile_discount.append(mobile_off[i].text)
```

```python
len(mobile_names),len(mobile_discount),len(mobile_rating)
```

```python
import pandas as pd
df=pd.DataFrame({'MOBILE_NAME':mobile_names,'DISCOUNT':mobile_discount,'RATING':mobile_rating})
df.head()
```

```python
df.to_excel('mobile.xlsx',index=False)
```
